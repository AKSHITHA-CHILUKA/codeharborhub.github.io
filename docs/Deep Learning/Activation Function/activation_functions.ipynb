<h3>Implementation of diff activation fucntion in neural network</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Non - Linaer activation functions
def sigmoid(x):
    return 1/(1+(np.exp(-x)))

def tanh(x):
    return((np.exp(x)) - np.exp(-x)) / ((np.exp(x)) + np.exp(-x))

def reLu(x):
    return np.maximum(x , 0)

# Linaer activation functions 
def identity(x):
    return x

def step(x , thres):
    return np.where (x >= thres , 1, 0)
```

```python
x = np.linspace(-10 , 10 , 200)
# plt.figure(figsize= (15 , 5))
plt.figure(figsize=(10, 6))

plt.plot(x , sigmoid(x) , label = "sigmoid" , color = "red")
plt.plot(x , tanh(x) , label = "Tan - h" , color = "blue")
plt.plot(x , reLu(x) , label = "reLu" , color = "green")
plt.plot(x , step(x , 3) , label = "Step" , color = "yellow")

plt.title("Activation functions")
plt.xlabel("input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
```
